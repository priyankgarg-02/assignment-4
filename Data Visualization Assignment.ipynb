## Q.2 
- Plot a Sine Wave with color code with "black" color and style as dashdot, also mark "x" and "y" label and give a suitable title to it

```python
x=np.linspace(0,12,1000)
plt.plot(x,np.sin(x),color='k',linestyle='-.')
plt.xlabel('x',color='blue',size='25')
plt.ylabel('y',color='red',size='25')
plt.title('Sine Wave',color='red',size='30')
```

## Q.4

-Plot Stacked Bar Graph for the above data.
    - Use proper colors and x and y label for the graph
    - Use legend positioned at upper right corner
    - Also mark xticks with an angle of 90 Degree

```python
store=['Delhi','Mumbai','Chennai','Kolkata','Hyderabad']
Clothing=np.array([82,78,48,46,39])
Equipment=np.array([48,31,51,29,18])
Accessories=np.array([15,20,14,13,9])
storeno=[x for x, _ in enumerate(store)]
plt.bar(storeno,Clothing,width=0.8,label='Clothing',color='magenta',bottom=Equipment+Accessories)
plt.bar(storeno,Equipment,width=0.8,label='Equipment',color='yellow',bottom=Accessories)
plt.bar(storeno,Accessories,width=0.8,label='Accessories',color='silver')
plt.xlabel('Store',size='25',color='blue')
plt.ylabel('Numbers of Goods',size='25',color='red')
plt.title('Store data',size='30',color='green')
plt.legend(loc='upper right')
plt.xticks(rotation=90)
plt.show()
```

## Q.5 For Titanic Dataset
- Load Dataset using Seaborn
- Plot a Bar Graph of Survival of Passenger against the Passenger Class.
- Use proper "hue" attribute, and palette color. Mark 'x' and 'y' label with proper title.
- Use legend located at upper left corner

```python
titanic=sns.load_dataset('titanic')
titanic.head(10)
```

```python
survive = sns.countplot(x = 'pclass', hue = 'survived', palette = sns.color_palette("muted",n_colors=2), data = titanic)
survive.set(title = 'Survival of Passenger against Passenger Class', xlabel = 'Passenger')
plt.legend(loc='upper right')
plt.show()
```

## Q.6 For Tips Dataset
- Load the Dataset Using Seaborn
- Plot the Horizontal Distribution Graph on Total Bill with color as Green
- Mark proper 'x' and 'y' label and give a suitable title

```python
tips=sns.load_dataset('tips')
tips.head(10)
```

```python
sns.set_style("darkgrid")
myxaxis=pd.Series(tips.total_bill,name='Total Bill')
grph=sns.distplot(myxaxis,color='green')
grph.set_title("Total Bill distribution Graph",size="20",color="blue")
```
